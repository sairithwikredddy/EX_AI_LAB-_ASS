```python
import os, sys, math, warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

DATA_PATH = "/content/kidney_disease.csv"
RANDOM_STATE = 42
TEST_SIZE = 0.2

if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(f"Dataset not found at {DATA_PATH}")
df = pd.read_csv(DATA_PATH)
print("Loaded:", DATA_PATH)
print("Shape:", df.shape)
print()

print("=== EDA ===")
print("Columns and dtypes:")
print(df.dtypes)
print("\nMissing value counts:")
print(df.isnull().sum())
print("\nFirst 5 rows:")
display(df.head())

print("\nSummary statistics (numeric):")
display(df.describe().T)

num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if num_cols:
    n = len(num_cols)
    cols = 3
    rows = math.ceil(n/cols)
    fig, axes = plt.subplots(rows, cols, figsize=(cols*4, rows*3))
    axes = axes.flatten()
    for i, c in enumerate(num_cols):
        axes[i].hist(df[c].dropna(), bins=20)
        axes[i].set_title(c)
    for j in range(i+1, len(axes)):
        axes[j].axis('off')
    plt.tight_layout()
    plt.show()
else:
    print("No numeric columns to plot.")

if len(num_cols) > 1:
    corr = df[num_cols].corr()
    plt.figure(figsize=(10,8))
    plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
    plt.colorbar()
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.index)), corr.index)
    plt.title("Correlation matrix (numerical features)")
    plt.tight_layout()
    plt.show()
else:
    print("Not enough numeric columns for correlation heatmap.")

possible_targets = [c for c in df.columns if c.lower() in ('target','class','diagnosis','disease','status','label','outcome')]
target_col = possible_targets[0] if possible_targets else df.columns[-1]
print("\nUsing target column:", target_col)
print("\nTarget value counts:")
display(df[target_col].value_counts(dropna=False))
plt.figure(figsize=(5,3))
df[target_col].value_counts().plot(kind='bar')
plt.title("Target counts")
plt.show()

print("\n=== Preprocessing ===")
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

data = df.copy()
data = data.replace('?', np.nan)

X = data.drop(columns=[target_col])
y_raw = data[target_col]

le_target = LabelEncoder()
y = le_target.fit_transform(y_raw.astype(str))
class_names = list(le_target.classes_)
n_classes = len(class_names)
print("Detected classes (n={}): {}".format(n_classes, class_names))

cat_cols = X.select_dtypes(include=['object','category']).columns.tolist()
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()

for c in X.columns:
    if c not in num_cols and c not in cat_cols:
        try:
            X[c] = pd.to_numeric(X[c])
            num_cols.append(c)
        except:
            cat_cols.append(c)

print("Numerical columns:", num_cols)
print("Categorical columns:", cat_cols)

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, num_cols),
    ('cat', categorical_transformer, cat_cols)
])

X_processed = preprocessor.fit_transform(X)
print("Processed feature shape:", X_processed.shape)

stratify_arg = y if len(np.unique(y)) > 1 else None
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=stratify_arg)
print("Train/test shapes:", X_train.shape, X_test.shape)
print("Train class distribution:", Counter(y_train))

try:
    from imblearn.over_sampling import SMOTE
    unique, counts = np.unique(y_train, return_counts=True)
    min_count = counts.min()
    if len(unique) > 1 and min_count > 1:
        k_neighbors = min(5, min_count - 1)
        sm = SMOTE(random_state=RANDOM_STATE, k_neighbors=k_neighbors)
        X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
        print("After SMOTE:", Counter(y_train_res), "(k_neighbors={})".format(k_neighbors))
    else:
        print("SMOTE skipped (too few samples in minority class or only one class).")
        X_train_res, y_train_res = X_train, y_train
except Exception as e:
    print("imblearn not available or SMOTE failed:", e)
    X_train_res, y_train_res = X_train, y_test

print("\n=== Machine Learning Models ===")
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report

models = {
    "LogisticRegression": LogisticRegression(max_iter=1000, random_state=RANDOM_STATE),
    "DecisionTree": DecisionTreeClassifier(random_state=RANDOM_STATE),
    "RandomForest": RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE),
    "SVM": SVC(probability=True, random_state=RANDOM_STATE),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "XGBoost": xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=RANDOM_STATE)
}

results = {}
predictions = {}
for name, model in models.items():
    try:
        model.fit(X_train_res, y_train_res)
        y_pred = model.predict(X_test)
        y_proba = None
        if hasattr(model, "predict_proba"):
            y_proba = model.predict_proba(X_test)
        else:
            try:
                dfun = model.decision_function(X_test)
                if dfun.ndim == 1:
                    y_proba = np.vstack([1-dfun, dfun]).T
                else:
                    y_proba = dfun
            except Exception:
                y_proba = None

        acc = accuracy_score(y_test, y_pred)
        prec_macro = precision_score(y_test, y_pred, average='macro', zero_division=0)
        rec_macro = recall_score(y_test, y_pred, average='macro', zero_division=0)
        f1_macro = f1_score(y_test, y_pred, average='macro', zero_division=0)
        roc = None
        if y_proba is not None and n_classes == 2:
            try:
                roc = roc_auc_score(y_test, y_proba[:,1])
            except Exception:
                roc = None
        elif y_proba is not None and n_classes > 2:
            try:
                roc = roc_auc_score(y_test, y_proba, multi_class='ovr')
            except Exception:
                roc = None

        results[name] = {"accuracy": acc, "precision_macro": prec_macro, "recall_macro": rec_macro, "f1_macro": f1_macro, "roc_auc": roc}
        predictions[name] = {"y_pred": y_pred, "y_proba": y_proba}
        print(f"\n{name} classification report:")
        print(classification_report(y_test, y_pred, target_names=class_names, zero_division=0))
    except Exception as e:
        print(f"Model {name} failed: {e}")

res_df = pd.DataFrame(results).T
print("\n=== ML Results Summary ===")
display(res_df)
res_df.to_csv("/content/ml_results_summary.csv")
print("Saved ML results to /content/ml_results_summary.csv")

for name, pred in predictions.items():
    print(f"\nConfusion matrix for {name}:")
    cm = confusion_matrix(y_test, pred["y_pred"])
    print(cm)

print("\n=== Deep Learning Models (MLP, 1D-CNN, LSTM, Autoencoder) ===")
try:
    import tensorflow as tf
    from tensorflow.keras.models import Sequential, Model
    from tensorflow.keras.layers import Dense, Input, Conv1D, Flatten, LSTM, MaxPooling1D, Reshape
    from tensorflow.keras.utils import to_categorical
    from tensorflow.keras.callbacks import EarlyStopping

    es = EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True)

    if n_classes == 2:
        y_train_dl = np.array(y_train_res)
        y_test_dl = np.array(y_test)
    else:
        y_train_dl = to_categorical(y_train_res, num_classes=n_classes)
        y_test_dl = np.array(y_test)

    n_features = X_train_res.shape[1]
    X_train_dl = X_train_res.reshape((-1, n_features, 1))
    X_test_dl = X_test.reshape((-1, n_features, 1))

    if n_classes == 2:
        final_activation = 'sigmoid'
        loss_fn = 'binary_crossentropy'
        out_units = 1
    else:
        final_activation = 'softmax'
        loss_fn = 'categorical_crossentropy'
        out_units = n_classes

    mlp = Sequential([Input(shape=(n_features,)),
                      Dense(64, activation='relu'),
                      Dense(32, activation='relu'),
                      Dense(out_units, activation=final_activation)])
    mlp.compile(optimizer='adam', loss=loss_fn, metrics=['accuracy'])
    mlp.fit(X_train_res, y_train_dl, validation_split=0.1, epochs=50, batch_size=32, callbacks=[es], verbose=0)
    if n_classes == 2:
        mlp_probs = mlp.predict(X_test).ravel()
        mlp_preds = (mlp_probs > 0.5).astype(int)
    else:
        mlp_probs = mlp.predict(X_test)
        mlp_preds = np.argmax(mlp_probs, axis=1)
    print("MLP classification report:")
    print(classification_report(y_test, mlp_preds, target_names=class_names, zero_division=0))

    cnn = Sequential([Input(shape=(n_features,1)),
                      Conv1D(32, kernel_size=3, activation='relu', padding='same'),
                      Conv1D(64, kernel_size=3, activation='relu', padding='same'),
                      Flatten(),
                      Dense(64, activation='relu'),
                      Dense(out_units, activation=final_activation)])
    cnn.compile(optimizer='adam', loss=loss_fn, metrics=['accuracy'])
    cnn.fit(X_train_dl, y_train_dl, validation_split=0.1, epochs=50, batch_size=32, callbacks=[es], verbose=0)
    if n_classes == 2:
        cnn_probs = cnn.predict(X_test_dl).ravel()
        cnn_preds = (cnn_probs > 0.5).astype(int)
    else:
        cnn_probs = cnn.predict(X_test_dl)
        cnn_preds = np.argmax(cnn_probs, axis=1)
    print("CNN classification report:")
    print(classification_report(y_test, cnn_preds, target_names=class_names, zero_division=0))

    lstm_model = Sequential([Input(shape=(n_features,1)),
                             LSTM(64),
                             Dense(32, activation='relu'),
                             Dense(out_units, activation=final_activation)])
    lstm_model.compile(optimizer='adam', loss=loss_fn, metrics=['accuracy'])
    lstm_model.fit(X_train_dl, y_train_dl, validation_split=0.1, epochs=50, batch_size=32, callbacks=[es], verbose=0)
    if n_classes == 2:
        lstm_probs = lstm_model.predict(X_test_dl).ravel()
        lstm_preds = (lstm_probs > 0.5).astype(int)
    else:
        lstm_probs = lstm_model.predict(X_test_dl)
        lstm_preds = np.argmax(lstm_probs, axis=1)
    print("LSTM classification report:")
    print(classification_report(y_test, lstm_preds, target_names=class_names, zero_division=0))

    hybrid = Sequential([Input(shape=(n_features,1)),
                        Conv1D(32, 3, activation='relu', padding='same'),
                        MaxPooling1D(pool_size=2),
                        LSTM(64),
                        Dense(32, activation='relu'),
                        Dense(out_units, activation=final_activation)])
    hybrid.compile(optimizer='adam', loss=loss_fn, metrics=['accuracy'])
    hybrid.fit(X_train_dl, y_train_dl, validation_split=0.1, epochs=50, batch_size=32, callbacks=[es], verbose=0)
    if n_classes == 2:
        hybrid_probs = hybrid.predict(X_test_dl).ravel()
        hybrid_preds = (hybrid_probs > 0.5).astype(int)
    else:
        hybrid_probs = hybrid.predict(X_test_dl)
        hybrid_preds = np.argmax(hybrid_probs, axis=1)
    print("Hybrid CNN+LSTM classification report:")
    print(classification_report(y_test, hybrid_preds, target_names=class_names, zero_division=0))

    input_layer = Input(shape=(n_features,))
    e = Dense(128, activation='relu')(input_layer)
    e = Dense(64, activation='relu')(e)
    bottleneck = Dense(16, activation='relu', name='bottleneck')(e)
    d = Dense(64, activation='relu')(bottleneck)
    d = Dense(128, activation='relu')(d)
    decoded = Dense(n_features, activation='linear')(d)
    auto = Model(inputs=input_layer, outputs=decoded)
    auto.compile(optimizer='adam', loss='mse')
    auto.fit(X_train_res, X_train_res, validation_split=0.1, epochs=50, batch_size=32, callbacks=[es], verbose=0)
    encoder = Model(inputs=input_layer, outputs=auto.get_layer('bottleneck').output)
    X_train_enc = encoder.predict(X_train_res)
    X_test_enc = encoder.predict(X_test)
    if n_classes == 2:
        clf_enc = Sequential([Input(shape=(X_train_enc.shape[1],)), Dense(32, activation='relu'), Dense(1, activation='sigmoid')])
        clf_enc.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        clf_enc.fit(X_train_enc, y_train_dl, validation_split=0.1, epochs=50, batch_size=32, callbacks=[es], verbose=0)
        enc_probs = clf_enc.predict(X_test_enc).ravel()
        enc_preds = (enc_probs > 0.5).astype(int)
    else:
        clf_enc = Sequential([Input(shape=(X_train_enc.shape[1],)), Dense(32, activation='relu'), Dense(n_classes, activation='softmax')])
        clf_enc.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
        clf_enc.fit(X_train_enc, y_train_dl, validation_split=0.1, epochs=50, batch_size=32, callbacks=[es], verbose=0)
        enc_probs = clf_enc.predict(X_test_enc)
        enc_preds = np.argmax(enc_probs, axis=1)
    print("Autoencoder+Classifier classification report:")
    print(classification_report(y_test, enc_preds, target_names=class_names, zero_division=0))

except Exception as e:
    print("TensorFlow or Keras not available or DL training failed:", e)
    print("DL models skipped. Run this section where TensorFlow is installed to execute DL models.")

print("\n=== XAI ===")
rf = models.get("RandomForest")
if rf is not None:
    try:
        importances = rf.feature_importances_
        feat_names = list(num_cols)
        if cat_cols:
            ohe = preprocessor.named_transformers_['cat'].named_steps['onehot']
            try:
                ohe_names = list(ohe.get_feature_names_out(cat_cols))
            except:
                ohe_names = list(ohe.get_feature_names(cat_cols))
            feat_names.extend(ohe_names)
        order = np.argsort(importances)[::-1]
        print("Top 15 RF features by importance:")
        for i in order[:15]:
            fname = feat_names[i] if i < len(feat_names) else f"f_{i}"
            print(f"{fname}: {importances[i]:.4f}")
    except Exception as e:
        print("RF feature importance failed:", e)

try:
    import shap
    if rf is not None:
        explainer = shap.TreeExplainer(rf)
        sample = X_test if X_test.shape[0] <= 200 else X_test[np.random.choice(range(X_test.shape[0]), 200, replace=False)]
        shap_values = explainer.shap_values(sample)
        shap.summary_plot(shap_values, sample, feature_names=feat_names, show=True)
    else:
        print("SHAP: RandomForest not found to explain.")
except Exception as e:
    print("SHAP not available or failed:", e)

try:
    from lime.lime_tabular import LimeTabularExplainer
    explainer_lime = LimeTabularExplainer(X_train_res, feature_names=feat_names, class_names=class_names, discretize_continuous=True)
    instance_idx = 0 if X_test.shape[0] > 0 else None
    if instance_idx is not None:
        exp = explainer_lime.explain_instance(X_test[instance_idx], rf.predict_proba, num_features=8)
        print("LIME explanation for one test instance:")
        print(exp.as_list())
except Exception as e:
    print("LIME not available or failed:", e)

print("\n=== Comparative Analysis (brief) ===")
print("ML results summary (macro metrics):")
display(res_df)
print("Notes:")
print("- Check ML vs DL reports above. DL models are more flexible but need more data & compute.")
print("- Tree models (RandomForest/XGBoost) typically provide best trade-off between accuracy and interpretability in tabular medical data.")
print("- Use SHAP/LIME/PDP to inspect important features (clinically review top features).")

print("\nPipeline finished. Outputs saved:")
print(" - ML results CSV: /content/ml_results_summary.csv")

plt.figure(figsize=(10, 6))
plt.bar(res_df.index, res_df['accuracy'], color='skyblue')
plt.xlabel("Machine Learning Model")
plt.ylabel("Accuracy")
plt.title("Machine Learning Model Accuracy Comparison")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
